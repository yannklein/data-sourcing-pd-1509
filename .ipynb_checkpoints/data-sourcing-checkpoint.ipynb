**Goal 3:** add the amount of LastFM listener per song from a LastFM DB dump

## 3️⃣ SQL

```python
import sqlite3
```

```python
limit = "500000"
```

```python
query = """
    SELECT ai.artist_mb, p.listeners_lastfm 
    FROM artist_info ai 
    JOIN popularity p ON ai.mbid = p.mbid 
    WHERE p.listeners_lastfm > ?
"""
```

```python
conn = sqlite3.connect('data/music.sqlite')
```

```python
# c = conn.cursor()
```

```python
# c.execute(query, (limit,))
```

```python
# c.fetchall()
```

```python
lastfm_df = pd.read_sql(query, conn, params=(limit,))
```

```python
lastfm_df.head(10)
```

```python
tracks_df = tracks_df.merge(
    lastfm_df,
    left_on='artists',
    right_on='artist_mb',
    how='left'
)
tracks_df.head(10)
```

**Goal 4:** add another track popularity info from BigQuery ListenBrainz DB

## 4️⃣ Google BigQuery

```python
!pip install --quiet pandas-gbq
```

```python
import pandas_gbq
```

```python
query = """
    SELECT artist_name, COUNT(recording_mbid) track_count
    FROM `lwt-data-lectures.listenbrainz.listen` 
    GROUP BY artist_name
    HAVING track_count > 1000
    ORDER BY track_count DESC
"""
```

```python
project_id='lwt-data-lectures'
```

```python
brainz_df = pandas_gbq.read_gbq(query, project_id=project_id)
```

```python
brainz_df.columns = ['artists', 'music_brainz_plays']
```

```python
tracks_df = tracks_df.merge(
    brainz_df,
    on='artists',
    how='left'
)
```

```python
tracks_df.head()
```

**Last Goal:** get the artist bday from Wikipedia!

## 5️⃣ Scraping

```python
import requests
from bs4 import BeautifulSoup
```

```python
def scrape(artist):
    formatted_artist = artist.replace(" ", "_")
    url = f'https://en.wikipedia.org/wiki/{formatted_artist}'
    try:
        html_doc = requests.get(url).content
        soup = BeautifulSoup(html_doc, 'html.parser')
        birthday = soup.find("span", class_='bday').text
        return birthday
    except:
        return 'Not found'
```

```python
scrape("Ed Sheeran")
```

```python
for index, row in tracks_df.iterrows():
    bday = scrape(row['artists'])
    tracks_df.loc[index, 'bday'] = bday
```

```python
tracks_df.head()
```

```python
tracks_df["music_brainz_plays"].hist(bins = 20);
```

```python
import re
pattern = r'\d{4}-\d{2}-\d{2}'
# Find only the rows that have valid birthdays
only_bdays = tracks_df[tracks_df["bday"].str.match(pattern)].copy()
# Convert to a datetime format
tracks_df["birthday"] = pd.to_datetime(only_bdays["bday"])
```
